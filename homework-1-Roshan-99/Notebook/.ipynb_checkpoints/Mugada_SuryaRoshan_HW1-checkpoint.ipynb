# 1 a ) Downloading the data set

```python
raw_data = loadarff('../Data/column_2C_weka.arff')
df = pd.DataFrame(raw_data[0])
df
```

```python
def convertClasses(label):
    #print(str(label))
    if label== b'Abnormal':
        return 1
    else:
        return 0
df['class']=df['class'].apply(lambda x:convertClasses(x))
df
```

## iii) Selecting train and test data set

```python
filt=(df['class']==0)
```

```python
train0=df[filt].iloc[0:70]
test0=df[filt].iloc[70:]
```

```python
train1=df[~filt].iloc[0:140]
test1=df[~filt].iloc[140:]
```

```python
train=pd.concat([train0,train1])
test=pd.concat([test0,test1])
```

```python
x_train,y_train=train.iloc[:,:-1],train['class']
x_test,y_test=test.iloc[:,:-1],test['class']
```

# 1 c) Classification using KNN

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

## i) Code for KNN using Euclidean distance

```python
#Basic KNN code
clf=KNeighborsClassifier(metric='euclidean')
clf.fit(x_train,y_train)
y_pred=clf.predict(x_test)
print('Test error',1-accuracy_score(y_test,y_pred))
```

## ii) Test different values of K ={208,205,...7,4,1}

```python

testErrorMap=dict()
trainErrorMap=dict()
decrementRate=-3
start=208
end=1
for k in range(start,end+decrementRate,decrementRate):
    clf=KNeighborsClassifier(n_neighbors=k)
    clf.fit(x_train,y_train)
    y_pred_test=clf.predict(x_test)
    y_pred_train=clf.predict(x_train)
    #print(accuracy_score(y_test,y_pred_test),accuracy_score(y_train,y_pred_train))
    testErrorMap[k]=1-accuracy_score(y_test,y_pred_test)
    trainErrorMap[k]=1-accuracy_score(y_train,y_pred_train)
```

```python
trainError=pd.DataFrame.from_dict(trainErrorMap,orient='index',columns=['train_error'])
```

```python
trainError
```

```python
testError=pd.DataFrame.from_dict(testErrorMap,orient='index',columns=['test_error'])
```

```python
testError
```

```python
errorMap=trainError.join(testError)
```

```python
sns.lineplot(data=errorMap)
plt.gca().invert_xaxis()
```

```python
minError=errorMap['test_error'].iloc[0]
optimalK=errorMap.index[0]
for index,row in errorMap.iterrows():
    val=row['test_error']
    if val<minError:
        minError=val
        optimalK=index
print(minError,optimalK)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import precision_recall_fscore_support
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
def printOutputMetrics(y_test,y_pred):
    confusion=confusion_matrix(y_test,y_pred)
    print('Confusion Matrix=\n',confusion)
    tn, fp, fn, tp =confusion.ravel()
    tpr=tp/(tp+fn)
    tnr=tn/(tn+fp)
    print('True positive rate=',tpr)
    print('True negative rate=',tnr)
    #precision,recall,f1_score,support=precision_recall_fscore_support(y_test,y_pred)
    print('Precision=',precision_score(y_test, y_pred))
    print('Recall=',recall_score(y_test, y_pred))
# f1_score(y_test, y_pred)
    print('F1 Score=',f1_score(y_test, y_pred))
```

```python

clf=KNeighborsClassifier(n_neighbors=optimalK)
clf.fit(x_train,y_train)
y_pred_test=clf.predict(x_test)
printOutputMetrics(y_test,y_pred_test)
```

## iii) Learning Rate

```python
def runKNN(traini,N,clf):
    start=1
    end=N
    increment=5
    bestError=1
    optimalK=start
    x_train,y_train=traini.iloc[:,:-1],traini['class']
    clf.n_neighbors=1
    clf.fit(x_train,y_train)
    y_predk1=clf.predict(x_test)
    errorK1=1-accuracy_score(y_test,y_predk1)
    for k in range(start+increment,end+1,increment):
        
        #x_test,y_test=test.iloc[:,:-1],test['class']
        #print(k,N,end+increment)
        #clf=KNeighborsClassifier(n_neighbors=k)
        clf.n_neighbors=k
        clf.fit(x_train,y_train)
        y_pred_test=clf.predict(x_test)
        error=1-accuracy_score(y_test,y_pred_test)
        if error <bestError:
            
            bestError=error
            optimalK=k
    #print(errorK1,1,bestError,optimalK)  
    if errorK1<bestError and bestError-errorK1>0.005:
            #print('returning 1')
            return errorK1,1
    #print('returning',optimalK)
    return bestError,optimalK
        
    
```

```python
#best test error rate or learning curve 
import math
start,end,increment=10,210,10
learningCurve=dict()
for n in range(start,end+increment,increment):
    
    train0i=train0[:math.floor(n/3)]
    train1i=train1[:math.floor(n-n/3)]
    traini=pd.concat([train0i,train1i])
    bestscore,K=runKNN(traini,n,KNeighborsClassifier())
    learningCurve[n]=bestscore
```

```python
learningCurve
```

```python
learningCurveDF=pd.DataFrame.from_dict(learningCurve,orient='index',columns=['Best Error Rate'])
```

```python
sns.lineplot(data=learningCurveDF)
plt.ylabel('Best Error Rate')
plt.xlabel('Training Size')
```

# 1 D) Using different distance metrics 

## i) Minkowski
### a) Manhattan

```python
#D part i
kNeighborsManhattan=KNeighborsClassifier(metric='manhattan')
bestErrorManhattan,optimalKManhattan=runKNN(train,196,kNeighborsManhattan)
kNeighborsManhattan.n_neighbors=optimalKManhattan
print('Optimal K=',optimalKManhattan)
#kNeighborsManhattan.fit(x_train,y_train)
#print(classification_report(y_test,kNeighborsManhattan.predict(x_test)))
```

### b) Varying p value

```python
#D part ib
errorDict=dict()
for i in range(1,11):
    p_val=10**(0.1*i)
    #print(p)
    clf=KNeighborsClassifier(metric='minkowski',p=p_val,n_neighbors=optimalKManhattan)
    #print(clf.p)
    clf.fit(x_train,y_train)
    y_pred=clf.predict(x_test)
    errorDict[p_val]=1-accuracy_score(y_test,y_pred)
    
```

```python
d_part_ib_error=pd.DataFrame.from_dict(errorDict,orient='index',columns=['error'])
```

```python
errorDict
```

```python
bestP=0.1
minErrorMinkowski=1
for p,error in errorDict.items():
    if error<minErrorMinkowski:
        minErrorMinkowski=error
        bestP=p  
print('Best log10(p)=',math.log(bestP,10))
```

#### The lowest training error will always be with k=1 and it will be 0. Since when k=1, it will consider itself as the neighbour giving 100% accuracy

```python
minError=errorMap['train_error'].iloc[0]
optimalK=errorMap.index[0]
for index,row in errorMap.iterrows():
    val=row['train_error']
    if val<minError:
        minError=val
        optimalK=index
print('Minimum training error',minError,'with k=',optimalK)
```
