### c) Fitting a simple linear regression model to predict output. The outliers are identified by the data which is more than 4/n in the cooks distance. These are indicated by red. There is statistical significance between all the independent variables and dependent variables since they have p value less than 0.05.

```python
def get_cooks(cooks,x,y):
    outliers=dict()
    n=len(cooks)
    
    #print(x[0])
    #print(y[0])
    for i in range(0,n):
        if cooks[i]>=4/n:
            outliers[x[i]]=y[i]
    #print(outliers)
    return outliers
```

```python
coefficients=dict()
for column in df.columns[:-1]:
    model=sm.OLS(endog=df['PE'],exog=sm.add_constant(df[column])).fit()
    sns.regplot(x=column,y=df.columns[-1],data=df,line_kws={'color': 'black'})
    #sm.graphics.influence_plot(model)
    influence=model.get_influence()
    cooks=influence.cooks_distance
    cooks_points=get_cooks(cooks[0],df[column],df['PE'])
    print(cooks)
    #plt.show()
    #print(df[column])
    #print(cooks_points)
    dist = cooks[0]
    plt.scatter(cooks_points.keys(),cooks_points.values(),color='red')
    #plt.scatter(df[column],cooks[0])
    plt.show()
    print(model.params)
    coefficients[column]=[column,model.params[column]]
    print(model.summary())
```

### KNN regression. 
#### Best K value is around 4-5 and has been changing based on the random split of the data but is around that value. We can notice the error reducing as k increases and again increases as k increases with a low error point around 4-5 range. 

```python
def runKNN(x_train,y_train,x_test,y_test,title):
    
    trainError=dict()
    testError=dict()
    for k in range(1,101):
        kRegression=KNeighborsRegressor(n_neighbors=k)
        kRegression.fit(x_train,y_train)
        y_pred_test=kRegression.predict(x_test)
        y_pred_train=kRegression.predict(x_train)
        testError[1/k]=mean_squared_error(y_test,y_pred_test)
        trainError[1/k]=mean_squared_error(y_train,y_pred_train)
    
    bestK=min(testError, key=testError.get)
    minError=testError[bestK]
    print('Best K=',1/bestK,' with error=',minError)
    
    #print(testError.values())
    #print('train=',trainError.values())
    trainError=pd.DataFrame.from_dict(trainError,orient='index',columns=['train_error'])
    testError=pd.DataFrame.from_dict(testError,orient='index',columns=['test_error'])
    errorMap=trainError.join(testError)
    sns.lineplot(data=errorMap)
    plt.xlabel('1/k')
    plt.ylabel('Errors')
    plt.title(title)
    plt.show()
    
```

```python
#part i- 1 using only variables
runKNN(x_train[X_cols],y_train,x_test[X_cols],y_test,'Raw data')
```

```python
#part i- 1 using normalized
min_max_scaler=preprocessing.MinMaxScaler()
x_train_normalized=pd.DataFrame(min_max_scaler.fit_transform(x_train[X_cols]))
x_test_normalized=pd.DataFrame(min_max_scaler.transform(x_test[X_cols]))
runKNN(x_train_normalized,y_train,x_test_normalized,y_test,'Normalized data')
```

# 2 

### a) Sample size n is large and number of predictors are small. A flexible model is good in this case since a larger sample size (n) represents the population data better and a flexible model fits a larger sample size better given the number of predictors are small.

### b) Number of predictors are high (p) and the sample size (n) is small. If the sample size is small then having a flexible model will make it overfit that data. Additionally, having large number of predictors can lead to overfitting. Hence when the number of predictors is high and the sample size is small using a inflexible model is a better idea since it has generalizing properties. 

### c) The relationship between predictors and response is highly non-linear. When the response is highly non-linear it is better to choose a flexbile model since more flexible models can model non-linear relationships better. Inflexible models cannot model non-linear relationships or relationships of higher order. 

### d)Variance is high. When the variance is high then using a flexible model leads to overfitting. Hence, in this case using a inflexible model will maintain the generalizing properties and it will not be affected by the presence of outliers.
