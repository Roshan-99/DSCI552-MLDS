```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline as imbpipe
from sklearn.metrics import hamming_loss
from sklearn.metrics import accuracy_score
import numpy as np
from sklearn.pipeline import Pipeline
from sklearn.svm import LinearSVC
from sklearn.pipeline import make_pipeline
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
data=pd.read_csv('../Data/Frogs_MFCCs.csv')
data
```

```python
labels=['Family','Genus','Species']
```

```python
 X_train, X_test, y_train, y_test = train_test_split(data.iloc[:,:-4], data[labels], test_size=0.3)
```

```python
X_train
```

### 1) b) Exact match and hamming score/loss

#### Exact match : in multilabel multioutput classification, exact match sees whether all the labels are exactly matching the predicted labels. If yes then a score of 1 is given, else a score of 0 is given.

#### Hamming loss : In multilabel classification, hamming loss is calculated as the avg of the hamming distance between the predicted and true value. instead of a strict 1 or 0 policy, it counts number of mismatches between each label (this is same as hamming distance between 2 vectors). the average of this mismatch count is hamming loss.

#### hamming score is 1-hamming loss

```python
def my_hamming_loss(y_test,y_pred):
    outputs=[]
    labels=y_test.columns
    for label in labels:
        outputs.append(hamming_loss(y_test[label],y_pred[label]))
    return np.mean(outputs)

def my_exact_match(y_test,y_pred):
    y_pred=pd.DataFrame.from_dict(y_pred)
    return np.all(y_pred.to_numpy() == y_test.to_numpy(), axis=1).mean()
        
```

### 1) b) ii) SVM with guassian kernel

```python
y_pred=dict()
print('For Gaussian SVM :')
for label in labels:
    svc=SVC(kernel='rbf',decision_function_shape='ovr')
    paramgrid={'C':[10**i for i in range(-3,3)],'gamma':[i/10 for i in range(1,20)]}
    stratified_cv=StratifiedKFold(n_splits=10)
    clf=GridSearchCV(svc,param_grid=paramgrid,cv=stratified_cv)
    clf.fit(X_train,y_train[label])
    print('Label=',label,'Best score=',clf.best_score_,'Best params',clf.best_params_)
    y_pred[label]=clf.predict(X_test)
        
print('Exact match=',my_exact_match(y_test,y_pred))
print('Hamming Loss=',my_hamming_loss(y_test,y_pred))
```

### 1) b) iii) SVM with L1 penalized and standardizing the input. 

Using pipeline for standardizing and then training. Have to standardize correctly in cross validation since we should not standardize both the train and validate data. The standardizer should be fitted to the train data and the validate should just be transformed. Pipeline takes care of this with fit and predict

```python
y_pred=dict()
print('For L1 penalized SVM:')
for label in labels:
    linear_svc=LinearSVC(penalty='l1', loss = 'squared_hinge',dual=False,max_iter=1000000)
    pipe = Pipeline([
    ('scale', StandardScaler()),
    ('linear', linear_svc)])
    paramgrid=[{'linear__C':[10**i for i in range(-3,3)]}]
    stratified_cv=StratifiedKFold(n_splits=10)
    
    clf=GridSearchCV(pipe,param_grid=paramgrid,cv=stratified_cv)
    clf.fit(X_train,y_train[label])
    print('Label=',label,'Best score=',clf.best_score_,'Best params',clf.best_params_)
    y_pred[label]=clf.predict(X_test)
        
print('Exact match=',my_exact_match(y_test,y_pred))
print('Hamming Loss=',my_hamming_loss(y_test,y_pred))
```

### 1) b) iii) L1 Penalized SVM with smote. 

As previously, using pipeline of smote, standardizer and linearsvc. smote should be only done on the train data and not on the validate data in cross validation. This is taken care by the pipeline's fit and predict

```python
y_pred=dict()
print('For L1 penalized SVM with Smote:')
for label in labels:
    linear_svc=LinearSVC(penalty='l1', loss = 'squared_hinge',dual=False,max_iter=1000000)
    pipe = imbpipe([
    ('smote',SMOTE()),
    ('scale', StandardScaler()),
    ('linear', linear_svc)])
    paramgrid=[{'linear__C':[10**i for i in range(-3,3)]}]
    stratified_cv=StratifiedKFold(n_splits=10)
    
    clf=GridSearchCV(pipe,param_grid=paramgrid,cv=stratified_cv)
    clf.fit(X_train,y_train[label])
    print('Label=',label,'Best score=',clf.best_score_,'Best params',clf.best_params_)
    y_pred[label]=clf.predict(X_test)
        
print('Exact match=',my_exact_match(y_test,y_pred))
print('Hamming Loss=',my_hamming_loss(y_test,y_pred))
```

### Conclusions: Gaussian SVM (better than)> L1 penalized (better than)> L1 penalized + smote  in both exact match as well as hamming loss

## Monte Carlo simulation 


```python
cluster_data=data.iloc[:,:-1]
X,y=cluster_data.iloc[:,:-3],cluster_data[labels]
```

```python
y
```

```python
def cluster_hamming_loss(y_test,y_pred):
    outputs=[]
    labels=y_pred.keys()
    #print(labels)
    for label in labels:
        #print(y_test[label],y_pred[label])
        outputs.append(hamming_loss([y_test[label]],[y_pred[label]]))
    return np.mean(outputs)

def cluster_hamming_distance(y_test,y_pred):
    outputs=[]
    labels=y_pred.keys()
    #print(labels)
    for label in labels:
        #print(y_test[label],y_pred[label])
        outputs.append(hamming_loss([y_test[label]],[y_pred[label]]))
    return np.sum(outputs)

def my_exact_match(y_test,y_pred):
    y_pred=pd.DataFrame.from_dict(y_pred)
    return np.all(y_pred.to_numpy() == y_test.to_numpy(), axis=1).mean()
        
```

```python
def find_best_k():
    range_n_clusters=range(2,5)
    best_K,best_score=1,0
    for n_clusters in range_n_clusters:
        clusterer = KMeans(n_clusters=n_clusters,init='random')
        preds = clusterer.fit_predict(X)
        centers = clusterer.cluster_centers_

        score = silhouette_score(X, preds)
        if score>best_score:
            best_score=score
            best_K=n_clusters
        #print("For n_clusters = {}, silhouette score is {}".format(n_clusters, score))
    #print('Best K=',best_K)
    return best_K,best_score
        
```

### 2) a, b,c 

```python
monte_hamming_distance=[]
for i in range(0,50):
    print('Iteration ',i,":")
    best_K,best_score=find_best_k()
    print('Optimal K=',best_K,' with score=',best_score)
    clusterer = KMeans(n_clusters=best_K)
    preds = clusterer.fit_predict(X)
    centers = clusterer.cluster_centers_

    pred_labels=dict()
    for k in range(best_K):
        pred_labels[k]=dict()
        for label in labels:
            pred_labels[k][label]=dict()


    for i in range(len(preds)):
        k=preds[i]
        for label in labels:
            true_label=data.iloc[i][label]
            pred_labels[k][label][true_label]=pred_labels[k][label].get(true_label,0)+1

    majority_labels=dict()
    for k in pred_labels:
        majority_labels[k]=dict()
        for label in labels:
            majority,major_class=0,'wrong'

            for key in pred_labels[k][label]:
                val=pred_labels[k][label][key]
                if val>majority:
                    majority=val
                    major_class=key

            majority_labels[k][label]=major_class

    print('Majority labels=',majority_labels)
    hamming_loss_arr=[]
    hamming_distance_arr=[]
    for i in range(len(data)):
        hamming_loss_arr.append(cluster_hamming_loss(data[labels].iloc[i],majority_labels[preds[i]]))
        hamming_distance_arr.append(cluster_hamming_distance(data[labels].iloc[i],majority_labels[preds[i]]))

    avg_hamming_loss=np.mean(hamming_loss_arr)
    avg_hamming_distance=np.mean(hamming_distance_arr)
    print('Avg hamming loss=',avg_hamming_loss)
    print('Avg hamming distance=',avg_hamming_distance)
    print('Avg hamming score',1-avg_hamming_loss)
    print('--------------------------------------------')
    monte_hamming_distance.append(avg_hamming_distance)
```

### 2) Monte carlo results

```python
print('Avg of hamming distance of 50 runs =',np.mean(monte_hamming_distance))
print('Standard deviation of hamming distance of 50 runs =',np.std(monte_hamming_distance))
```
